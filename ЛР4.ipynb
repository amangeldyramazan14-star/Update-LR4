```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

```python
# Pipeline с логистической регрессией
pipeline_lr = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression(max_iter=2000, random_state=42))
])

# обучение
pipeline_lr.fit(X_train, y_train)

# предсказание
y_pred_lr = pipeline_lr.predict(X_test)

# accuracy
acc_lr = accuracy_score(y_test, y_pred_lr)

print("Accuracy Logistic Regression:", acc_lr)
```

```python
# GridSearch
param_grid = {
    "model__C": [0.1, 1, 10],
    "model__solver": ["lbfgs"]
}

grid = GridSearchCV( pipeline_lr, param_grid, cv=5, scoring="accuracy" )

# обучение
grid.fit(X_train, y_train)

print("Лучшие параметры:", grid.best_params_)
```

```python
#Лучшая модель
best_lr = grid.best_estimator_

y_pred_best = best_lr.predict(X_test)
acc_best = accuracy_score(y_test, y_pred_best)

print("Accuracy после GridSearch:", acc_best)
```

```python
#Альтернативная модель Random Forest
rf = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf.fit(X_train, y_train)

y_pred_rf = rf.predict(X_test)
acc_rf = accuracy_score(y_test, y_pred_rf)

print("Accuracy Random Forest:", acc_rf)
```
